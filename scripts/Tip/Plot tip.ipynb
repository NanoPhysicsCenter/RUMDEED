```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from scipy.constants import pi, hbar, e, m_e
```

```python
# System

V_0 = 1000
time_step = 0.25E-15

# Fixed shape

R_r = 250.0E-9
h_r = 500.0E-9 # [m]
h = h_r
d = 1000.0E-9 # [m]
d_plane = d + h_r

xi_r = h_r/d + 1
a = np.sqrt(d**2*R_r**2/(h_r**2+2*d*h_r) + d**2)
eta_1 = - d / a

shift_zd = h - h_r
max_xi = - (h+d)/(a*eta_1)

theta = np.arccos(d/a)
r_tip = a*np.sin(theta)*np.tan(theta)
eta_2 = 0.0 # Defines the plane of absorption
eta = eta_1 # Defines the tip
shift_z = np.abs(a*eta*xi_r)

Q_0 = 1/2*np.log((1+np.cos(theta))/(1-np.cos(theta)))
r = a*np.sin(theta)*np.tan(theta)
```

```python
def f(x, y):
    return eta_1/np.sqrt(1.0 - eta_1**2) * np.sqrt(x**2 + y**2 + a**2*(1.0 - eta_1**2)) + shift_z

x = np.linspace(-6, 6, 30)
y = np.linspace(-6, 6, 30)

X, Y = np.meshgrid(x, y)
Z = f(X, Y)
```

```python
fig = plt.figure()
ax = plt.axes(projection='3d')
ax.plot_surface(X, Y, Z, rstride=1, cstride=1,
                cmap='viridis', edgecolor='none')
```
