```python
# IMPORTS

import numpy as np
import pandas as pd

import math as m

import matplotlib.pyplot as plt

import seaborn as sns
sns.set_theme()
```

<h3>Simulations

```python
ratios = [0, 2, 6]

filebase = './R'
filebase2 = '/out/ramo_current.dt'

plt.figure(figsize=(12, 6)) 
for ratio in ratios:
    filename = filebase + str(ratio) + filebase2
    df = pd.read_csv(filepath_or_buffer=filename, index_col = 1,delim_whitespace=True, \
                    header=None, names=['time', 'step', 'current', 'volt', 'nrPart', 'nrElec', 'nrIon', 'avg_mob', 'avg_part_speed','avg_elec_speed','avg_ion_speed','a','b','c'])
    # Plot distribution of kramers radii
    plt.plot(df.index, df['nrIon'], label=f'Ratio = 0.{ratio}')
tot_nr_atom = (df["nrIon"][1])/(ratios[-1]/10)
print(f'Total number of atoms: {tot_nr_atom}')
plt.title('Number of ions in the system for different intitial ratios.')
plt.xlabel('Times step')
plt.ylabel('Number of ions')
plt.ylim(0,tot_nr_atom)
plt.legend()
plt.show()
```
