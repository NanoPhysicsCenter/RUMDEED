```python
# IMPORTS

import numpy as np
import pandas as pd

import math as m

import matplotlib.pyplot as plt

import seaborn as sns
sns.set_theme()
```

<h3>Simulations

```python
fontz = 26

ratios = [0, 2, 6]

filebase = './R'
filebase2 = '/out/ramo_current.dt'

plt.figure(figsize=(18, 8)) 
for ratio in ratios:
    filename = filebase + str(ratio) + filebase2
    df = pd.read_csv(filepath_or_buffer=filename, index_col=1, delim_whitespace=True, \
                     header=None, names=['time', 'step', 'current', 'volt', 'nrPart', 'nrElec', 'nrIon', 'avg_mob', 'avg_part_speed','avg_elec_speed','avg_ion_speed','a','b','c'])
    # Plot distribution of kramers radii
    plt.plot(df.index*1.0e-4, df['nrIon'], label=f'Ratio = 0.{ratio}')
    print(np.mean(df['nrIon'][400000:500000]), np.mean(df['nrIon'][500000:600000]))
tot_nr_atom = (df["nrIon"][1]) / (ratios[-1] / 10)
print(f'Total number of atoms: {tot_nr_atom}')

plt.xticks(fontsize=fontz)
plt.yticks(fontsize=fontz)
plt.xlabel('Time [ps]', fontsize=fontz)
plt.ylabel('Number of ions', fontsize=fontz)
plt.ylim(0,500)
plt.legend(fontsize=fontz)
plt.gcf().set_facecolor('white')
plt.gca().set_facecolor('white')
plt.grid(True)
plt.show()
```
